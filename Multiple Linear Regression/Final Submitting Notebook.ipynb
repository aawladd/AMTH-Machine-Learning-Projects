```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Real estate.csv')
df.head(10)
```

```python
df.info()
```

```python
df.corr()['Y house price of unit area']
```

```python
df = df.drop(["No","X1 transaction date"], axis =1)
df
```

```python
df_np = df.to_numpy()
df_np.shape
```

```python
X, y = df_np[:, :3], df_np[:, -1]
X.shape, y.shape
```

```python
from numpy.linalg import inv

def get_best_model(X, y):

    (n, p_minus_one) = X.shape
    p = p_minus_one + 1

    new_X = np.ones(shape=(n, p))
    new_X[:, 1:] = X

    return np.dot(np.dot(inv(np.dot(new_X.T, new_X)), new_X.T), y)
```

```python
b = get_best_model(X,y)
b
```

```python
def get_predictions(model, X):


    (n, p_minus_one) = X.shape
    p = p_minus_one + 1

    new_X = np.ones(shape=(n, p))
    new_X[:, 1:] = X

    return np.dot(new_X, model)
```

```python
predictions = get_predictions(b,X)
predictions, predictions.shape
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mean_absolute_error(predictions,y ), mean_squared_error(y, predictions)
```
