<h1>OLS Regression in Matrix Form</h1>

$y = Xb+e$</b>

$ e = y - Xb$

$\sum e_i^{2} = e'e$

$e'e = (y-Xb)'(y-Xb)$
$    = y'y - 2b'X'y+b'X'Xb$<br>

$\frac{\partial e'e}{\partial b} = -2X'y + 2X'Xb=0$<br>

$\therefore (X'X)b=X'y$ <br>

$\therefore b = (X'X)^{-1}X'y$
